```python
url='https://en.wikipedia.org/wiki/2022_FIFA_World_Cup'
```

```python
response=requests.get(url)
response
```

* Beautiful Soup then parses the document using the best available parser.
* The prettify() method will turn a Beautiful Soup parse tree into a nicely formatted Unicode string, 
 with a separate line for each tag and each string

```python
soup=BeautifulSoup(response.text,"html.parser")
print(soup.prettify())
```

* soup.find_all method will return the content from the tag we want.
* 'div' tag is generic container for a block of HTML.
* The 'div' tag has class which is an attribute used to define a group of HTML element.
* The content we want is stored inside this tag so we'll work on this.
* Here we have inspected the information of the group A matches from Wikipedia web.
* On this basis we're going to find the (date,time,matches) that are going to took place in fifaworldcup2022.

```python
tags=soup.find_all("div",{"class":"footballbox"})
```

```python
tags
```

* Here we created empty list which will be storing our scrapped data.
* with using for loop we'll go through every tags of our data
* The tags are  used soup.find() method because there are only one tag in the data which has our output
* These tags are generated by inspecting sections from the web

```python
# Date=[]
# Time=[]
# Home_team=[]
# Away_team=[]
# Matches11=[]

for tag in tags:
    date=tag.find("span",{"class":"bday dtstart published updated"})
    print(date.get_text())
    time=tag.find("div",{"class":"ftime"})
    print(time.get_text())
    home_team=tag.find("th",{"itemprop":"homeTeam"})
    print(home_team.get_text())
    away_team=tag.find("th",{"itemprop":"awayTeam"})
    print(away_team.get_text())
    matches11=home_team.get_text(strip=True) + " vs " + away_team.get_text(strip=True)
    print(matches11)
```

* Here we have append our data into the empty list we have created
* we have appended our data into the list using the tags only
* To access human-readable text we used inisdd a tag a method called get_text()


```python
Date=[]
Time=[]
Home_team=[]
Away_team=[]
Matches1=[]
l=0

for tag in tags:
    Date.append(tag.find("span",{"class":"bday dtstart published updated"}).get_text())
    Time.append(tag.find("div",{"class":"ftime"}).get_text())
    Home_team.append(tag.find("th",{"itemprop":"homeTeam"}).get_text())
    Away_team.append(tag.find("th",{"itemprop":"awayTeam"}).get_text())
    Matches1.insert(l ,tag.find("th",{"itemprop":"homeTeam"}).get_text()+ " vs " + tag.find("th",{"itemprop":"awayTeam"}).get_text())
    l+=1
    
   
                           
        
     
    
```

* We have printed a appended list just to check the output

```python
print(Matches1)
```

* Here we have created a dictionary called data which will have column name as key and list as value
* meaning we gave the list values a key

```python
Data={'Date':Date,'Time':Time,'Home_team':Home_team,'Away_team':Away_team,'Matches':Matches1}
```

```python
Data
```

* Using pandas we have dataframed and analyzed our data

```python
Fifa=pd.DataFrame(Data)
Fifa
```

* Finally we saved our data in csv foramat
* A CSV File (Comma separated values file) is type of plain text file that uses specific structuring to arrange tabular data.
* We can view our csv file data in excelsheet
